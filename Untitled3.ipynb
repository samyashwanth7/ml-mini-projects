```python
# Install & import
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Load dataset directly
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/sms.tsv"
df = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])

# Encode labels
df['label_num'] = df.label.map({'ham': 0, 'spam': 1})

# Split & vectorize
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label_num'], test_size=0.2)
cv = CountVectorizer()
X_train_cv = cv.fit_transform(X_train)
X_test_cv = cv.transform(X_test)

# Train model
model = MultinomialNB()
model.fit(X_train_cv, y_train)

# Accuracy
print("Accuracy:", accuracy_score(y_test, model.predict(X_test_cv)))
```

```python
# Test YOUR OWN messages
def predict(msg):
    result = model.predict(cv.transform([msg]))
    return "🚨 SPAM" if result[0] == 1 else "✅ HAM (Not Spam)"

print(predict("Congratulations! You won a free for nothing just pranking bro"))
print(predict("Hey bro, what time is the class tomorrow?"))
```

```python
# See what words scream SPAM to your model
import numpy as np

feature_names = cv.get_feature_names_out()
spam_probs = model.feature_log_prob_[1]  # spam class
top_spam_words = [feature_names[i] for i in np.argsort(spam_probs)[-15:]]
print("Most SPAM words:", top_spam_words)
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load data directly - no URL needed!
data = fetch_20newsgroups(subset='all', categories=['rec.sport.hockey', 'sci.space'], remove=('headers', 'footers'))

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2)

tfidf = TfidfVectorizer(max_features=5000)
X_train_t = tfidf.fit_transform(X_train)
X_test_t = tfidf.transform(X_test)

model = LogisticRegression()
model.fit(X_train_t, y_train)
print("Accuracy:", accuracy_score(y_test, model.predict(X_test_t)))
```

```python
def predict_topic(text):
    result = model.predict(tfidf.transform([text]))
    return "🏒 Hockey" if result[0] == 1 else "🚀 Space"

print(predict_topic("NASA launched a new rocket to the moon today"))
print(predict_topic("The goalkeeper saved an amazing penalty shot"))
print(predict_topic("Virat Kohli scored a century"))  # 👀 see what happens lol
```
